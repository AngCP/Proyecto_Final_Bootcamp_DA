### 10. Most customers are interested in properties with three or four bedrooms. What is the difference in average prices of the properties with three and four bedrooms? In this case you can simply use a `groupby` to check the prices for those particular houses.

```python
# Precio medio de las casas con 3 y 4 habitaciones:
df3 = house_price_data.groupby('bedrooms').mean()
media_3 = df3.loc[3, 'price']
print('Media 3 habitaciones:', media_3)

df4 = house_price_data.groupby('bedrooms').mean()
media_4 = df4.loc[4, 'price']
print('Media 4 habitaciones:', media_4)

# Diferencia de precio:
diferencia = media_4 - media_3
print('Diferencia:', diferencia)
```
